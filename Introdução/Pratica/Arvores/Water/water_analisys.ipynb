```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
water_data_path = './water_potability.csv'
water_data = pd.read_csv(water_data_path)
water_data.info()
```

```python
water_data.describe()
```

```python
water_data['Potability'].value_counts()
```

```python
water_data['Potability'].replace(
    {0: 'Não potável',
     1: 'Potável'}
)
```

```python
water_data['Potability'].value_counts()
```

```python
water = water_data.dropna()
```

```python
water.info()
```

```python
y = water.iloc[:,9]
y
```

```python
X = water.iloc[:,0:9]
X
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
clf = tree.DecisionTreeClassifier(random_state=1)
```

```python
clf.fit(train_X, train_y)
```

```python
predicao = clf.predict(val_X)
```

```python
score = accuracy_score(val_y, predicao)
print(f"Score: {score:.2%}")
```
