```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
X_full = pd.read_csv('../train.csv', index_col='Id')
X_full.head(5)
```

```python
X_test_full = pd.read_csv('../test.csv', index_col='Id')
X_test_full.head(5)
```

```python
X_test_full.shape
```

```python
X_test_full.info()
```

```python
y = X_full.SalePrice
```

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = X_full[features].copy()
X_test = X_test_full[features].copy()
```

```python
X
```

```python
X_test
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
#Definindo os modelos
model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=100, criterion='absolute_error', random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
model_5 = RandomForestRegressor(n_estimators=100, max_depth=7, random_state=0)

models = [model_1, model_2, model_3, model_4, model_5]
```

```python
from sklearn.metrics import mean_absolute_error
#Função para comparar os diferentes modelos usando a métrica MAE
def score_model(model, X_t = X_train, X_v = X_valid, y_t = y_train, y_v = y_valid):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)
for i in range(0, len(models)):
    mae = score_model(models[i])
    print(f"Modelo {i + 1}: MAE {mae}")
```

```python
meu_modelo = RandomForestRegressor(random_state=0)
meu_modelo.fit(X, y)
```

```python
predicao_teste = meu_modelo.predict(X_test)
```

```python
output = pd.DataFrame({'Id': X_test.index, 'SalePrice': predicao_teste})
output
```
